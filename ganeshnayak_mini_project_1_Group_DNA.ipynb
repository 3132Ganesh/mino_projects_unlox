```python
from datetime import datetime

def parse_chat(filename):
    chat_data = []
    with open(filename, "r", encoding="utf-8") as f:
        for line in f:

            if " - " in line:
                try:
                    timestamp_str, rest = line.split(" - ", 1)

                    dt = datetime.strptime(timestamp_str.strip(), "%d/%m/%y, %H:%M")
                    if ": " in rest:
                        sender, message = rest.split(": ", 1)
                        chat_data.append({
                            "timestamp": dt,
                            "sender": sender.strip(),
                            "message": message.strip(),
                            "type": "text"
                        })
                    else:

                        chat_data.append({
                            "timestamp": dt,
                            "sender": None,
                            "message": rest.strip(),
                            "type": "system"
                        })
                except ValueError:

                    continue
    return chat_data

parsed = parse_chat("hostel_bois.txt")
for entry in parsed[:10]:  # Shows first 10 entries
    print(entry)
```

```python
def grp_overview(chat_data):
  total_mssg=len(chat_data)
  participants=set()
  per_person_count={}
  for e in chat_data:
    sender=e['sender']
    if sender:
      participants.add(sender)
      per_person_count[sender]=per_person_count.get(sender,0)+1
  start_D=chat_data[0]["timestamp"]
  end_D=chat_data[-1]["timestamp"]
  print("===================================================\n GROUP OVERVIEW \n===================================================")
  print(f"total mssgs :{total_mssg}")
  print(f" participants :{len(participants)}->\nNAMES:{','.join(participants)}")
  print(f"start date  :{start_D}  \nEnd date  :{end_D}\n")
  print("MESSAGE PER PERSON")
  for sender,count in per_person_count.items():
    print(f"{sender} : {count}")
  return participants
```

```python
from collections import Counter
from datetime import datetime
def most_active_day_hr(chat_data):
  dayscount=Counter()
  hrscount=Counter()
  for e in chat_data:
    ts = e['timestamp']
    dayscount[ts.date()]+=1
    hrscount[ts.hour]+=1

  if not dayscount:
      print("No valid timestamps found to determine active days/hours.")
      return

  busiest_day=max(dayscount,key=dayscount.get)
  busiest_hr=max(hrscount,key=hrscount.get)
  print("Most Active Day & hour")
  print(f"Busiest day:{busiest_day}({dayscount[busiest_day]} message)")

  print(f"Busiest hour :{busiest_hr} :00 ({hrscount[busiest_hr]} messages)")
most_active_day_hr(parsed)
```

```python
from collections import Counter
from datetime import datetime

def top_5_days_hours(chat_data):
    day_counts = Counter()
    hour_counts = Counter()

    for entry in chat_data:
        ts = entry["timestamp"] # ts is now guaranteed to be a datetime object
        # Removed datetime.strptime, as it's no longer needed and would cause a TypeError
        day_counts[ts.date()] += 1
        hour_counts[ts.hour] += 1

    # Top 5 busiest days
    top_days = day_counts.most_common(5)
    # Top 5 busiest hours
    top_hours = hour_counts.most_common(5)

    print(" Top 5 Busiest Days")
    print("Date         | Messages")
    print("-----------------------")
    for day, count in top_days:
        print(f"{day} | {count}")

    print("\n Top 5 Busiest Hours")
    print("Hour   | Messages")
    print("------------------")
    for hour, count in top_hours:
        print(f"{hour:02d}:00 | {count}")
top_5_days_hours(parsed)
```

```python
import numpy as np
from datetime import datetime

def activity_heatmap(chat_data):
    participants = sorted({e["sender"] for e in chat_data if e["sender"]})
    heatmap = np.zeros((len(participants), 24), dtype=int)

    for e in chat_data:
        sender = e["sender"]
        ts = e["timestamp"]
        if sender:

            hour = ts.hour
            row = participants.index(sender)
            heatmap[row, hour] += 1

    print("ACTIVITY HEATMAP (messages by hour)")
    print("Hour:      " + " ".join([f"{h:02d}" for h in range(24)]))

    for i, sender in enumerate(participants):
        max_val = heatmap[i].max()
        row_visual = []
        for val in heatmap[i]:
            if max_val == 0:
                row_visual.append(". ")
            else:
                ratio = val / max_val
                if ratio == 0:
                    row_visual.append("  ")
                elif ratio <= 0.25:
                    row_visual.append("░  ")
                elif ratio <= 0.50:
                    row_visual.append("▒  ")
                elif ratio <= 0.75:
                    row_visual.append("▓  ")
                else:
                    row_visual.append("█  ")
        print(f"{sender:10}| {''.join(row_visual)}")
activity_heatmap(parsed)
```

```python
import string
def top_words(chat_data,top_n=10):
  stop_words = {"i","is","the","a","and","or","to","of","in","on","for","you","me","it","my","we","was","this","how","about","at","am","so","he","she","his","just","which","today","that","up","have","one","are","from","had","they","no"}
  word_count={}
  for e in chat_data:
    word=e["message"].lower().split()
    for w in word:
      w=w.strip(string.punctuation)
      if w and w not in stop_words:
        word_count[w]=word_count.get(w,0)+1
  sorted_words=sorted(word_count.items(),key=lambda x:x[1],reverse=True)[:top_n]
  print("THIS GROUP'S FRVT WORDS")
  for word,count in sorted_words:
    bar="‖"*(count//5)
    print(f"{word:<10} {bar} {count}")
  return word_count
word_count = top_words(parsed)
```

```python
def response_stats(chat_data):
    response_times = {}
    silent_streaks = {}

    prev_time = None
    prev_sender = None

    for entry in chat_data:
        sender = entry["sender"]
        ts = entry["timestamp"]

        if sender and isinstance(ts, datetime):
            # Response time calculation
            if prev_time and sender != prev_sender:
                diff = (ts - prev_time).total_seconds() / 60  # minutes
                response_times.setdefault(sender, []).append(diff)

            # Silent streak calculation
            if prev_time:
                diff = (ts - prev_time).total_seconds() / 60
                silent_streaks[sender] = max(silent_streaks.get(sender, 0), diff)

            prev_time = ts
            prev_sender = sender

    print(" Response Speed (avg minutes)")
    avg_response_times = {}
    for sender, times in response_times.items():
        avg = sum(times) / len(times)
        avg_response_times[sender] = avg
        print(f"{sender:10} | {avg:.2f} min")

    if avg_response_times:
        fastest_replier = min(avg_response_times, key=avg_response_times.get)
        slowest_replier = max(avg_response_times, key=avg_response_times.get)
        print(f"\nFastest Replier: {fastest_replier} ({avg_response_times[fastest_replier]:.2f} min)")
        print(f"Slowest Replier: {slowest_replier} ({avg_response_times[slowest_replier]:.2f} min)")

    print("\n Longest Silent Streak")
    for sender, streak in silent_streaks.items():
        if streak >= 1440:  # more than a day
            days = streak / 1440
            print(f"{sender:10} | {days:.1f} days")
        elif streak >= 60:  # more than an hour
            hours = streak / 60
            print(f"{sender:10} | {hours:.1f} hours")
        else:
            print(f"{sender:10} | {streak:.1f} min")
    return response_times, silent_streaks
response_times, silent_streaks = response_stats(parsed)
```

```python
def detect_archetypes(chat_data, word_counts, response_times, silent_streaks, heatmap, participants):
    archetypes = {}

    # Spammer: highest message count
    msg_counts = {p: sum(heatmap[i]) for i, p in enumerate(participants)}
    spammer = max(msg_counts, key=msg_counts.get)
    archetypes[spammer] = "Spammer"

    # Ghost: longest silent streak
    ghost = max(silent_streaks, key=silent_streaks.get)
    archetypes[ghost] = "Ghost"

    # Group Mom/Dad: most caring words
    caring_words = {"please", "everyone", "guys", "anyone"}
    caring_scores = {p: sum(word_counts.get(w, 0) for w in caring_words) for p in participants}
    mom = max(caring_scores, key=caring_scores.get)
    archetypes[mom] = "Group Mom/Dad"

    # Night Owl: most messages after 10
    night_scores = {p: sum(heatmap[i][22:]) for i, p in enumerate(participants)}
    night_owl = max(night_scores, key=night_scores.get)
    archetypes[night_owl] = "Night Owl"

    # Early Bird: most messages before 08
    early_scores = {p: sum(heatmap[i][:8]) for i, p in enumerate(participants)}
    early_bird = max(early_scores, key=early_scores.get)
    archetypes[early_bird] = "Early Bird"

    # Print results
    print(" Group Archetypes")
    for p in participants:
        role = archetypes.get(p, "Balanced Contributor")
        print(f"{p:10} → {role}")
detect_archetypes(parsed, word_count, response_times, silent_streaks, heatmap, participants_from_heatmap)
```
